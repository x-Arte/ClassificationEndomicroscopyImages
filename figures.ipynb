```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
filename='model/acc/2023-12-12-18-04.csv'

df = pd.read_csv(filename)

plt.title('accuracy')
plt.xlabel('epoch')
plt.plot(df['epoch'], df['train accuracy'],label = 'Training accuracy',linewidth=0.5 )
plt.plot(df['epoch'], df['test accuracy'],label = 'Testing accuracy',linewidth=0.5 )
plt.legend()
# inset
ax_inset = plt.axes([0.5, 0.4, 0.3, 0.3])  
ax_inset.plot(df['epoch'], df['train accuracy'], label='Training Accuracy', linewidth=0.5)
ax_inset.plot(df['epoch'], df['test accuracy'], label='Testing Accuracy', linewidth=0.5)
ax_inset.set_title('Zoomed in')
ax_inset.set_xlim([70, 150])  
ax_inset.set_ylim([98.5, 100])
plt.show()



plt.title('loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.plot(df['epoch'], df['training loss'],label = 'Training loss',linewidth=0.5 )
plt.plot(df['epoch'], df['testing loss'],label = 'Testing loss',linewidth=0.5 )
plt.legend()
# inset
ax_inset = plt.axes([0.5, 0.4, 0.3, 0.3])  
ax_inset.plot(df['epoch'], df['training loss'], label='Training loss', linewidth=0.5)
ax_inset.plot(df['epoch'], df['testing loss'], label='Testing loss', linewidth=0.5)
ax_inset.set_title('Zoomed in')
ax_inset.set_xlim([110, 150])  
ax_inset.set_ylim([0, 0.08])
plt.show()

plt.title('AUC')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.plot(df['epoch'], df['AUC score'],label = 'AUC score',linewidth=0.5 )
plt.show()
```
